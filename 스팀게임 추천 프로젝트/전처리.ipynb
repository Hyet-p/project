## 데이터 불러오기

```python
import sqlite3
import pandas as pd

# 데이터베이스에 연결
conn = sqlite3.connect('steamDB')

# 쿼리를 실행하여 데이터를 가져옵니다.
df = pd.read_sql_query("SELECT * from steam_app_data", conn)

conn.close()  # 데이터베이스 연결을 종료합니다.

# 데이터 확인
df.head()
```

## 데이터 전처리

### 전체적 전처리

```python
# 필요한 열만 선택
df_selected = df[['name','is_free','price_overview','categories','genres','detailed_description','developers','publishers','platforms','release_date']]
df_selected.head()
```

```python
# info 확인
df_selected.info()
```

```python
# 공백인 데이터 제거
df_selected = df_selected[df_selected['name'] != '']
df_selected = df_selected[df_selected['is_free'] != '']
df_selected = df_selected[df_selected['categories'] != '']
df_selected = df_selected[df_selected['genres'] != '']

# 유료인데 가격이 나와있지 않은 데이터 제거 
df_selected = df_selected[(df_selected['is_free'] != 'False') | (df_selected['price_overview'] != '')]
```

```python
# 게임 이름을 기준으로 중복 제거
df_selected = df_selected.drop_duplicates(subset='name')
```

```python
df_selected.reset_index(drop=True, inplace=True)
df_selected.info()  # 1597개 데이터 제거됨
```

```python
# price_overview가 비어있는 건 다 무료이므로 0으로 변경
df_selected['price_overview'] = df_selected['price_overview'].replace('', 0)
df_selected.head()
```

### 열별 전처리

#### price_overview

```python
df_selected['price_overview'][0]    # 마지막 final_formatted 값인 £7.19를 가져와서, 원화로 바꿀 것
```

```python
import ast

# 가격 추출, 원화(￦)로 바꾸는 함수
def extract_price(value):
    if value == 0:
        return 0
    else:
        try:
            dictionary = ast.literal_eval(value)
            final_formatted = dictionary['final_formatted']
            if '£' in final_formatted:
                price = float(final_formatted.replace('£', '')) * 1600
            elif '$' in final_formatted:
                price = float(final_formatted.replace('$', '')) * 1300
            else:
                price = 0
            return price
        except:
            return None
    
# 'price' 열 생성
df_selected['price'] = df_selected['price_overview'].apply(extract_price)

# price_overview 열 드랍
df_selected = df_selected.drop('price_overview', axis=1)

# 예외가 발생한 행 삭제
df_selected = df_selected.dropna()
df_selected.reset_index(drop=True, inplace=True)
```

```python
df_selected.head()
```

```python
# 이상값 확인
df_selected[df_selected['price']>200000]
```

```python
# 스팀에서 확인해본 결과, 아래 게임들은 자리수가 잘못된 것 확인
df_selected.loc[11136, 'price'] *= 0.01
df_selected.loc[13313, 'price'] *= 0.5

correct_price_idx = [10846, 15201, 17468, 21516]
df_selected.loc[correct_price_idx, 'price'] *= 0.1

df_selected[df_selected['price']>200000]
```

#### categories

```python
df_selected['categories'][0]    # description의 값들을 ;로 연결하여 가져올 것
```

```python
# 'description' 값 추출 함수
def extract_category(value):
    try:
        category_list = ast.literal_eval(value)
        descriptions = [category['description'] for category in category_list]
        return ';'.join(descriptions)
    except Exception as e:
        return str(e)

# 'category' 열 생성
df_selected['category'] = df_selected['categories'].apply(extract_category)

# categories 열 드랍
df_selected = df_selected.drop('categories', axis=1)
```

```python
df_selected.head()
```

#### genres

```python
df_selected['genres'][0]    # description의 값들을 ;로 연결하여 가져올 것
```

```python
# 'description' 값 추출 함수
def extract_genre(value):
    try:
        genre_list = ast.literal_eval(value)
        descriptions = [genre['description'] for genre in genre_list]
        return ';'.join(descriptions)
    except Exception as e:
        return str(e)

# 'genre' 열 생성
df_selected['genre'] = df_selected['genres'].apply(extract_genre)

# genres 열 드랍
df_selected = df_selected.drop('genres', axis=1)
```

```python
df_selected.head()
```

#### platforms

```python
df_selected['platforms'][0]    # 값이 True인 키값들을 ;로 연결하여 가져올 것
```

```python
# True인 키값 추출 함수
def extract_platforms(value):
    try:
        platform_list = ast.literal_eval(value)
        descriptions = [k for k, v in platform_list.items() if v]
        return ';'.join(descriptions)
    except Exception as e:
        return e

# 'platform' 열 생성
df_selected['platform'] = df_selected['platforms'].apply(extract_platforms)

# platforms 열 드랍
df_selected = df_selected.drop('platforms', axis=1)
```

```python
df_selected.head()
```

#### release_date

```python
df_selected['release_date'][0]    # date의 값을 YYYY-MM-DD 형태로 가져올 것
```

```python
from datetime import datetime

# release_date 추출 함수
def extract_release_date(value):
    try:
        dictionary = ast.literal_eval(value)
        date_str = dictionary['date']
        date_obj = datetime.strptime(date_str, '%d %b, %Y')
        return date_obj.date()
    except Exception as e:
        return e

# 'release_date' 열 업데이트
df_selected['release_date'] = df_selected['release_date'].apply(extract_release_date)
```
